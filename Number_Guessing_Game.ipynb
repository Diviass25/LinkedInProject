### Number Guessing Game Requirements

ğŸ² The program should pick a random number between 1 and 100.

ğŸ‘¤ The user must try to guess the number.

ğŸ“¢ After each guess, the program should say if the guess is too high or too low.

ğŸ”„ The user should have several chances to guess the number.

ğŸ† The game should end immediately if the user guesses correctly.

â³ The game should also end after the maximum number of attempts if the user doesnâ€™t guess it.

```python
import random

# Generate a random number between 1 and 100
target_number = random.randint(1, 100)

# Allow a maximum of 10 guesses
max_attempts = 10
attempts = 0

print("Welcome to the Number Guessing Game!")
print(f"I have picked a number between 1 and 100. You have {max_attempts} attempts.")

# Loop continues as long as the number hasn't been guessed and user has attempts left
while attempts < max_attempts:
    try:
        guess = int(input(f"Attempt {attempts + 1}/{max_attempts}: Enter your guess: "))
        attempts += 1

        if guess < 1 or guess > 100:
            print("Please guess a number between 1 and 100.")
        elif guess < target_number:
            print("Too low! Try again.")
        elif guess > target_number:
            print("Too high! Try again.")
        else:
            # The output if the user has successfully guessed the number
            print(f"Congratulations! You guessed the number {target_number} in {attempts} attempts!")
            break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

# The output if the user didn't guess the number correctly
if attempts == max_attempts and guess != target_number:
    print(f"Sorry, you've run out of attempts. The number was {target_number}.")
print("Thanks for playing!")
```
