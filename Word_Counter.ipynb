### Word Count Program Requirements

üìÇ The program should **open and read a text file**.  
‚úÇÔ∏è It should **split the file content into words**.  
üî¢ Then, it should **count how many words are in the file**.  
‚ö†Ô∏è If the file is missing or cannot be opened, the program must **show an error message** instead of crashing.  

```python
def count_words_in_file(filepath):
    """
    Reads a text file and counts the number of words in it.

    Args:
        filepath (str): The path to the text file.

    Returns:
        int: The total number of words in the file, or -1 if the file is not found.
    """
    try:
        with open(filepath, 'r') as file:
            content = file.read()
            words = content.split()
            return len(words)
    except FileNotFoundError:
        print(f"Error: The file at '{filepath}' was not found.")
        return -1

# Example usage:
file_path = 'sample_data/README.md'  # Modified to use an existing file
word_count = count_words_in_file(file_path)

if word_count != -1:
    print(f"The file '{file_path}' contains {word_count} words.")
```
