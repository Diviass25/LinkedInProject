## Project Requirements
- Retrieve the Wikipedia page for Michael Jackson, and extract the page title and the first few paragraphs.
- Show the first three paragraphs in the output.

This code snippet demonstrates:
1.  **Importing Libraries**: `requests` for making web requests and `BeautifulSoup` for parsing HTML.
2.  **Fetching a Page**: Using `requests.get()` to retrieve the HTML content of a Wikipedia page.
3.  **Parsing HTML**: Creating a `BeautifulSoup` object from the page's content.
4.  **Extracting Information**: Finding and printing the page's title and a few paragraphs.

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the Wikipedia page you want to scrape
url = "https://en.wikipedia.org/wiki/Michael_Jackson"

# Add a User-Agent header to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send an HTTP GET request to the URL with headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the title of the page
    title = soup.find('h1', class_='firstHeading').text
    print(f"Page Title: {title}")

    # You can now start extracting other information from the 'soup' object
    # For example, let's find all the paragraphs
    paragraphs = soup.find_all('p')
    print(f"\nNumber of paragraphs found: {len(paragraphs)}")

    # Print the text of the first few paragraphs
    print("\nFirst 3 paragraphs:")
    for i, p in enumerate(paragraphs[:3]):
        print(f"Paragraph {i+1}: {p.text[:200]}...") # Print first 200 characters

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
