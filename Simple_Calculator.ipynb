### Interactive Calculator

```python
# Defining functions for addition, subtraction, multiplication, and division
def add(x, y):
    """Adds two numbers and returns the sum."""
    return x + y

def subtract(x, y):
    """Subtracts the second number from the first and returns the difference."""
    return x - y

def multiply(x, y):
    """Multiplies two numbers and returns the product."""
    return x * y

def divide(x, y):
    """Divides the first number by the second and returns the quotient. Handles division by zero."""
    if y == 0:
        return "Error: Cannot divide by zero!"
    return x / y
```

```python
# Reading two numbers from user
while True:
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        break
    except ValueError:
        print("Invalid input. Please enter numbers only.")
```

```python
# Reading an operation from user
while True:
    operation = input("Choose an operation (+, -, *, /): ")
    if operation in ('+', '-', '*', '/'):
        break
    else:
        print("Invalid operation. Please choose from +, -, *, /.")
```

```python
# Calculating two numbers with selected operation
if operation == '+':
    result = add(num1, num2)
elif operation == '-':
    result = subtract(num1, num2)
elif operation == '*':
    result = multiply(num1, num2)
elif operation == '/':
    result = divide(num1, num2)

print(f"Result: {result}")
```
