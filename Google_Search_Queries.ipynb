```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time

Trending_topics = TrendReq(hl='en-US', tz=360)
```

```python
kw_list=["Data Science"]
Trending_topics.build_payload(kw_list,cat=0, timeframe='today 12-m')
time.sleep(5)
```

```python
data = Trending_topics.interest_over_time()
display(data.head())
```

This table shows the search interest for 'Data Science' over time. The values are on a scale from 0 to 100, where 100 is the peak popularity for the term. A value of 0 means there was not enough data for this term.

Now, let's explore related queries and topics to understand what other terms people are searching for in relation to 'Data Science'.

```python
import time

try:
    time.sleep(10) # Add a delay to avoid TooManyRequestsError
    related_queries = Trending_topics.related_queries()
    keyword = kw_list[0] # Assuming only one keyword in kw_list
    if keyword in related_queries:
        if 'top' in related_queries[keyword] and not related_queries[keyword]['top'].empty:
            print("Top related queries:")
            display(related_queries[keyword]['top'])
        else:
            print("No top related queries found for this keyword.")

        if 'rising' in related_queries[keyword] and not related_queries[keyword]['rising'].empty:
            print("\nRising related queries:")
            display(related_queries[keyword]['rising'])
        else:
            print("No rising related queries found for this keyword.")
    else:
        print(f"No related queries data found for '{keyword}'.")

except Exception as e:
    print(f"An error occurred while fetching related queries: {e}")
```

The 'top' related queries show the most popular search terms related to 'Data Science', ranked by search volume. The 'rising' queries indicate terms that have had the most significant growth in search volume over the specified period, often reflecting emerging trends or increasing interest.

Let's also look at related topics.

```python
import time

try:
    time.sleep(10) # Add a delay to avoid TooManyRequestsError
    related_topics = Trending_topics.related_topics()
    keyword = kw_list[0] # Assuming only one keyword in kw_list
    if keyword in related_topics:
        # Display 'top' related topics
        if 'top' in related_topics[keyword] and not related_topics[keyword]['top'].empty:
            print("Top related topics:")
            display(related_topics[keyword]['top'])
        else:
            print("No top related topics found for this keyword.")

        # Display 'rising' related topics
        if 'rising' in related_topics[keyword] and not related_topics[keyword]['rising'].empty:
            print("\nRising related topics:")
            display(related_topics[keyword]['rising'])
        else:
            print("No rising related topics found for this keyword.")
    else:
        print(f"No related topics data found for '{keyword}'.")
except Exception as e:
    print(f"An error occurred while fetching related topics: {e}")
```

Similar to related queries, 'top' related topics show the most popular concepts and entities related to 'Data Science', while 'rising' related topics highlight those with significant recent growth in interest.
