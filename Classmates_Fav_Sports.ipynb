## Project Requirements
Analyze the distribution of classmates' favorite sports. This includes generating a synthetic dataset, calculating the frequency and percentage of each sport, visualizing the distribution with a bar chart, and summarizing the findings.

## Generate Dataset

Generate a synthetic dataset of classmates' favorite sports, including columns for 'Classmate ID' and 'Favorite Sport'.


```python
import pandas as pd
import numpy as np

print("Libraries imported successfully.")
```

```python
sports = ['Basketball', 'Soccer', 'Swimming', 'Tennis', 'Volleyball', 'Track & Field', 'Cycling', 'Baseball', 'Badminton', 'Gymnastics']
classmate_ids = np.arange(1, 101) # 100 classmates

# Assign different probabilities to sports to create a 'statistically meaningful distribution'
# For example, making Basketball and Soccer more popular, and Gymnastics less popular.
probabilities = [0.20, 0.20, 0.10, 0.10, 0.10, 0.08, 0.07, 0.07, 0.05, 0.03]
# Ensure probabilities sum to 1
if not np.isclose(sum(probabilities), 1.0): # Use np.isclose for floating point comparison
    print("Warning: Probabilities do not sum to 1. Adjusting them to sum to 1.")
    probabilities = np.array(probabilities) / sum(probabilities)

favorite_sports = np.random.choice(sports, size=len(classmate_ids), p=probabilities)

# Create the DataFrame
df_sports = pd.DataFrame({
    'Classmate ID': classmate_ids,
    'Favorite Sport': favorite_sports
})

print("Synthetic dataset generated successfully.")
# Display the first few rows of the DataFrame
print(df_sports.head())
```

## Calculate Frequency and Percentage

Calculate the frequency (count) and percentage distribution of each favorite sport among the classmates.

```python
sport_counts = df_sports['Favorite Sport'].value_counts()
total_classmates = len(df_sports)
sport_percentages = (sport_counts / total_classmates) * 100

df_sport_distribution = pd.DataFrame({
    'Frequency': sport_counts,
    'Percentage': sport_percentages
})

print("Distribution of Favorite Sports (Frequency and Percentage):")
print(df_sport_distribution)
```

## Visualize Distribution

Create a bar chart to visualize the frequency or percentage distribution of favorite sports.


```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Matplotlib and Seaborn imported successfully.")
```

```python
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=df_sport_distribution.index, y='Percentage', data=df_sport_distribution, hue=df_sport_distribution.index, legend=False, palette='viridis')

plt.title('Distribution of Favorite Sports Among Classmates', fontsize=16)
plt.xlabel('Favorite Sport', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Add percentage values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points', fontsize=9)

plt.tight_layout()
plt.show()

print("Bar chart visualizing sport distribution generated successfully.")
```

### Data Analysis Key Findings

*   A synthetic dataset of 100 classmates and their favorite sports was successfully generated, incorporating 10 diverse sports with varied probabilities to simulate a statistically meaningful distribution.
*   The frequency and percentage of each sport preference were calculated, revealing a clear hierarchy of popularity.
*   **Basketball** emerged as the most popular sport, favored by 26.0% of classmates.
*   **Soccer** followed as the second most popular, preferred by 19.0% of classmates.
*   **Volleyball** ranked third with 11.0% of preferences.
*   **Badminton** and **Tennis** were identified as the least popular sports, both preferred by only 3.0% of classmates.
*   The distribution was visualized using a bar chart, clearly illustrating the percentage breakdown for each sport and highlighting the disparities in popularity.
